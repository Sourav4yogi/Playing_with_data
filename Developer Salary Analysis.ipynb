```python
fig=plt.figure(figsize=(10,6))
plt.plot(ages,salary,color='grey',linestyle='--',marker='o',label='Average Developer')
plt.plot(ages,pydevsal,color='Red',marker='o',label='Python Developer')
plt.plot(ages,jsdevsal,color='Yellow',marker='o',label='JavaScript Developer')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary by Age')
```
