```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp=pd.read_csv('../input/daily-temperature-of-major-cities/city_temperature.csv')
```

```python
indtemp=temp[temp['Country']=='India']
indtemp['City'].unique()
```

```python
indtemp=indtemp.replace('Bombay (Mumbai)','Mumbai')
mumtemp=indtemp[indtemp['City']=='Mumbai']
mumtemp=mumtemp.drop(['State'],axis=1)
deltemp=indtemp[indtemp['City']=='Delhi']
deltemp=deltemp.drop(['State'],axis=1)
chtemp=indtemp[indtemp['City']=='Chennai (Madras)']
chtemp=chtemp.drop(['State'],axis=1)
caltemp=indtemp[indtemp['City']=='Calcutta']
caltemp=caltemp.drop(['State'],axis=1)
caltemp
```

```python
mumyear=mumtemp.groupby(['Year']).mean()
delyear=deltemp.groupby(['Year']).mean()
chyear=chtemp.groupby(['Year']).mean()
calyear=caltemp.groupby(['Year']).mean()
calyear
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(18,9))
plt.plot(mumyear.index,mumyear['AvgTemperature'],label='Mumbai')
plt.plot(delyear.index,delyear['AvgTemperature'],label='Delhi')
plt.plot(chyear.index,chyear['AvgTemperature'],label='Chennai')
plt.plot(calyear.index,calyear['AvgTemperature'],label='Kolkata')
plt.xticks(mumyear.index)
plt.xlabel('Years')
plt.ylabel('Average Annual Temperature')
plt.title('Annual temperatures of Indian Metros from 1995 to 2020')
plt.legend()
```
