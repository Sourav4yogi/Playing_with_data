```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:/Users/soura/Desktop/Datasets/gapminder.csv')
df.head()
```

```python
us=df[df['country']=='United States']
china=df[df['country']=='China']
india=df[df['country']=='India']
years=us['year']
```

```python
plt.figure(figsize=(12,8))
plt.style.use('seaborn')
plt.plot(years,us['pop'],color='blue',marker='o',label='United States')
plt.plot(years,china['pop'],color='red',marker='o',label='China')
plt.plot(years,india['pop'],color='green',marker='o',label='India')
plt.xlabel('Years')
plt.ylabel('Population in 100M')
plt.title('Population Growth')
plt.legend()
```

```python
newdf=df[df['year']==2007]
rich=newdf.sort_values(by='gdpPercap',ascending=False).head(20)
rich=rich.sort_values(by='gdpPercap')
```

```python
plt.figure(figsize=(12,8))
plt.barh(rich['country'],rich['gdpPercap'])
plt.xlabel('GDP per Capita in USD')
plt.title('20 Richest Countries in the World(2007)')
```

```python
newdf['GDP']=newdf['pop']*newdf['gdpPercap']
```

```python
newrich=newdf.sort_values(by='GDP',ascending=False).head(20)
newrich=newrich.sort_values(by='GDP')
```

```python
plt.figure(figsize=(12,8))
plt.barh(newrich['country'],newrich['GDP'])
plt.xlabel('GDP in USD')
plt.title('20 Richest Countries in the World by overall GDP(2007)')
```

```python
plt.figure(figsize=(10,6))
bins=list(range(30,100,10))
plt.hist(newdf['lifeExp'],edgecolor='k',bins=bins,log=False)
plt.xlabel('Life expectency')
plt.ylabel('No of Countries')
plt.axvline(newdf['lifeExp'].median(),color='red',label='Median Life expectency')

plt.legend()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(newdf['gdpPercap'],newdf['lifeExp'],s=100,edgecolor='k',linewidth=1,alpha=0.75,cmap='summer',color='r')
```
