```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median 
```

```python
iris=pd.read_csv('../input/iris-flower-dataset/IRIS.csv')
iris.head()
```

```python
sns.set_style('darkgrid')
```

```python
plt.figure(figsize=(13,6.5))
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species',size='petal_length')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(13,6.5))
sns.histplot(data=iris,x='petal_width',bins=20,hue='species',multiple='stack')
```

```python
plt.figure(figsize=(13,6.5))

sns.barplot(data=iris,x='species',y='petal_length',estimator=median,palette="Blues_d")
```

```python
plt.figure(figsize=(13,6.5))
sns.boxplot(data=iris,x='species',y='sepal_width')
```

```python
plt.figure(figsize=(13,6.5))
sns.boxplot(data=iris,orient='h',palette='Set3')
```

```python
plt.figure(figsize=(13,6.5))
sns.heatmap(data=iris.corr(),annot=True)
```

```python
sns.pairplot(data=iris,hue='species')
```
